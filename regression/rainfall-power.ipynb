### Linear Regression: Daily Rainfall with Power Demand

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/daily-weather-power/dhaka/2022-01-01-2022-12-31.csv')
df
```

```python
# Create scatter plot for Power Demand vs Rainfall
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='rainfall', y='power_demand')
plt.title('Power Demand vs Rainfall')
plt.xlabel('Rainfall')
plt.ylabel('Power Demand')
plt.show()
```
